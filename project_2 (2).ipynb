#Linear Regression on Simple Dataset
###This graph visualizes the relationship between TV advertising spend (in ₹000s) and sales (also in ₹000s) using simple linear regression.

#**Approach** 1:

```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```
# Step 1: Load dataset
data = {
    'TV': [50, 80, 100, 120, 150, 170, 200, 220, 250, 270],
    'Sales': [7, 9, 12, 15, 18, 21, 25, 26, 30, 32]
}
df = pd.DataFrame(data)
```

```
# Step 2: Prepare data
X = df[['TV']]  # Feature must be 2D
y = df['Sales']
```

```
# Step 3: Train model
model = LinearRegression()
model.fit(X, y)
```

```
# Step 4: Print model details
print("Coefficient (Slope):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```
# Step 5: Predict and visualize
y_pred = model.predict(X)
```

```
plt.scatter(X, y, color='blue', label="Actual")
plt.plot(X, y_pred, color='red', label="Regression Line")
plt.xlabel("TV Advertising Spend (₹k)")
plt.ylabel("Sales (₹k)")
plt.title("Simple Linear Regression: TV vs Sales")
plt.legend()
plt.grid(True)
plt.show()
```

## **📊 Insight Summary**
The graph shows a positive linear relationship between TV ad spend and sales. As companies spend more on TV advertising, sales generally increase.

The red line is the regression line, showing predicted sales.

The blue dots are actual sales, and they mostly align well with the line — indicating a good model fit.

Small deviations suggest other factors may also affect sales.

## **Conclusion:**
TV advertising has a clear, positive impact on sales, based on this simple linear regression model.

#**Approach 2 :**

```
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```
# Step 1 : Dataset
data = {
    'TV': [250, 180, 90, 120, 60, 200, 130, 100, 50, 220],
    'Radio': [30, 20, 50, 35, 25, 40, 45, 30, 15, 50],
    'Social_Media': [55, 65, 40, 60, 20, 55, 35, 45, 10, 60],
    'YouTube': [40, 25, 10, 30, 15, 50, 40, 35, 20, 60],
    'Sales': [27, 21, 15, 22, 10, 25, 19, 18, 8, 30]
}

df = pd.DataFrame(data)
```

```
# Step 2 : Features and target
X = df[['TV', 'Radio', 'Social_Media', 'YouTube']]
y = df['Sales']
```

```
# Step 3 : Model
model = LinearRegression()
model.fit(X, y)
```

```
# Step 4 : Coefficients and intercept
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```
# Step 5 : Predict values
y_pred = model.predict(X)
```

```
# Step 6 : Plotting Actual vs Predicted
plt.scatter(y, y_pred, color='teal')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # Reference line
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted Sales")
plt.grid(True)
plt.show()
```

#📊 Quick Insight:
###Multiple Linear Regression — Actual vs Predicted Sales
This graph shows how well the linear regression model predicts sales using 4 features: TV, Radio, Social Media, and YouTube.

##Key Points:
The red dashed line is the ideal prediction (where predicted = actual).

The teal dots are real model predictions.

Dots are closely aligned to the red line → model is predicting sales accurately.
